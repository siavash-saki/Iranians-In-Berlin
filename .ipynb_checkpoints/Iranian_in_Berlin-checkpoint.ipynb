### 1.1 Data Wrangling

```python
# create a path for each file
dataset_path_1 = os.path.join(os.getcwd(), 'dataset1')
dataset_names_1 = os.listdir(dataset_path_1)
dataset_path_names_1 = [os.path.join(dataset_path_1, name) for name in dataset_names_1 ]
dataset_path_names_1.sort()
```

```python
# number of files
len(dataset_names_1)
```

```python
# files names
sorted(dataset_names_1)
```

```python
# read files
dataset_1=[]
for file in dataset_path_names_1:
    dataset_1.append(pd.read_excel(file))
```

```python
# download tables from Berlin wikipedia
berlin_wiki=pd.read_html('https://de.wikipedia.org/wiki/Verwaltungsgliederung_Berlins')
berlin_Bezirk= berlin_wiki[1]
berlin_Ortsteil= berlin_wiki[2]
```

```python
berlin_Bezirk.head()
```

```python
# create a df with Bezirk and ID for each Bezirk
berlin_Bezirk.rename(columns={'Nr.': 'Bezirk_ID'},inplace=True)
berlin_IDs= berlin_Bezirk[['Bezirk_ID','Bezirk']]
berlin_IDs= berlin_IDs.append({'Bezirk_ID':'0','Bezirk':'Berlin'}, ignore_index=True)
```

```python
# modify name of Bezirk and ID for each df in dataset
dataset_1= [df.drop('Bezirk',axis=1) for df in dataset_1]
dataset_1= [pd.concat([berlin_IDs,df],axis=1) for df in dataset_1]
```

Create the final dataframe to do the analysis:

```python
# features df
features=[df.iloc[:,2:] for df in dataset_1[:-1]]
features=pd.concat(features,axis=1)
# target df
target=dataset_1[15]
# concat these two
df_Bezirk= pd.concat([target,features],axis=1)
# drop last row
df_Bezirk.drop(index=12,inplace=True)
```

```python
df_Bezirk
```

### 1.2 Analysis

```python
from scipy import stats

# calculate pearson coef and p values to show the relation
pearson_coef, p_value = stats.pearsonr(df_Bezirk['Iranians_In_Berlin'], df_Bezirk['Alter_Auslaender Insgesamt'])

print(f'Correlation Coefficient= {pearson_coef}, p-value= {p_value}')
```
